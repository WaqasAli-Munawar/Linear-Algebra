At its core, a matrix is a way to represent a table of numbers. All of the matrices we worked with in the last project contained 2 rows and 3 columns. Here's the first one that setup our linear system:



This is known as a $2x3$ matrix (pronounced "two by three matrix"). The convention in linear algebra is to specify the number of rows first (2) then the number of columns (3).

Each of the rows and columns in this matrix is represented as a list of numbers. A list of numbers is known as a **vector**. A row from a matrix is known as a **row vector**, while a column is known as a **column vector**.

In this project, we'll learn more about column **vectors** and their associated operations to help us understand certain properties of linear systems.

We'll start by building some geometric intuition of vectors. Generally, the word vector refers to the column vector (ordered list of elements in a single column)

Whenever a vector contains 2 or 3 elements, we can visualize it on a coordinate grid. Typically, a vector is visualized on a coordinate grid using arrows, not using coordinates. For a vector containing two elements, the first value describes the x coordinate while the second value describes the y coordinate. Vectors are usually drawn from the origin (0,0) to the point described by the vector: 

Arrows are used to visualize individual vectors because they emphasize the 2 key properties of a vector -- direction and magnitude. The direction of a vector describes the way its pointing while the magnitude describes its length

We can visualize vectors in matplotlib using the [pyplot.quiver()](https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.quiver) function. This function has 4 required parameters: **X, Y, U, V**. 

X and Y correspond to the  coordinates we want the vector to start at and U and V correspond to the  coordinates we want the vector to end at

We also need to set the angles and scale_units parameters to xy and the scale parameter to 1. Setting angles to 'xy' lets matplotlib know we want the angle of the vector to be between the points we specified. The scale_units and scale parameters lets us specify custom scaling parameters for the vectors. When plotting vectors, we always want to use these 3 values for the parameters:

plt.quiver(0, 0, 1, 2, angles='xy', scale_units='xy', scale=1)

Lastly, we can set the color of each vector using the color parameter and passing in a string.

```python
import matplotlib.pyplot as plt
import numpy as np

plt.axhline(0,c = "black", lw = 0.5)
plt.axvline(0, c = "black", lw = 0.5)
plt.xlim(-3,3)
plt.ylim(-4,4)

# Generate a vector from [0,0] to [2,3] and set its color to 'blue'.
plt.quiver(0,0,2,3, angles = "xy" ,scale_units = "xy", scale = 1, color = "blue")

# Generate a vector from [0,0] to [-2,-3] and set its color to 'blue'.
plt.quiver(0,0,-2,-3,angles = "xy", scale_units = "xy", scale = 1, color = "blue")

# Generate a vector from [0,0] to [1,1] and set its color to 'gold'
plt.quiver(0, 0, 1, 1, angles='xy', scale_units='xy', scale=1, color='gold')

# Generate a vector from [0,0] to [2,2] and set its color to 'gold'.
plt.quiver(0, 0, 2, 2, angles='xy', scale_units='xy', scale=1, color='gold')

plt.show()
```

Like the rows in a matrix, vectors can be added and subtracted together. When we add or subtract vectors, elements in the same position are added or subtracted and a new vector is created. Here's an example of 2 vectors being added:


Here's an example of vector subtraction, using the same 2 vectors:



To visualize vector addition, we need to connect the vectors together instead of having all of the vectors originate at (0,0). 

```python
# This code draws the x and y axis as lines.
plt.axhline(0, c='black', lw=0.5)
plt.axvline(0, c='black', lw=0.5)
plt.xlim(-4,4)
plt.ylim(-1,4)

# Plot the vector [3,0]. This vector should originate at (0,0) .
plt.quiver(0,0,3,0,angles = "xy", scale_units = "xy", scale = 1)

# Plot the vector [0,3]. This vector should originate at (3,0) .
plt.quiver(3,0,0,3,angles = "xy", scale_units = "xy", scale = 1)

#Add both of these vectors and plot the resulting vector
plt.quiver(0,0,3,3,angles = "xy", scale_units = "xy", scale = 1, color = "green")
```

Between vector addition, vector subtraction, scalar multiplication, and scalar division, we can combine vectors in many different ways. Being able to combine vectors to create other vectors is a key idea in linear algebra. For example, we can combine the following vectors:

```python
plt.axhline(0, c='black', lw=0.5)
plt.axvline(0, c='black', lw=0.5)
plt.xlim(0,10)
plt.ylim(0,5)
plt.quiver(0, 0, 3, 1, angles='xy', scale_units='xy', scale=1, color='blue')
plt.quiver(0, 0, 6, 2, angles='xy', scale_units='xy', scale=1, color='green')
plt.quiver(0, 0, 9, 3, angles='xy', scale_units='xy', scale=1, color='orange')
```

To represent vectors in NumPy, we use the same function but we only specify a single value in each row. 

```python
import numpy as np

vector_one = np.asarray([
    [1],
    [2],
    [1]
], dtype=np.float32)
vector_two = np.asarray([[3],
                        [0],
                        [1]], dtype = np.float32)

print(vector_one)
print()
print(vector_two)
```

As expected, we can use the + and the - operators to perform vector addition and subtraction. In addition, we can use the * operator to multiply a vector by a scalar quantity.

```python
vector_linear_combination = (2*vector_one) +( 5 * vector_two)
vector_linear_combination
```

The last critical vector operation is the dot product. Dot product is similar to multiplication, but involves multiplying the elements in the vectors in a specific way. Both of these vectors need to contain the same number of elements a requirement the other operations as well.

To compute the dot product, we need to sum the products of the 2 values in each position in each vector 

Unlike the other vector operations, the result of a dot product is a scalar value, not a vector. To compute the dot product between 2 vectors, we need to use the numpy.dot() function. This function accepts 2 NumPy ndarray objects as the 2 required parameters. The main quirk is that one of the two vectors need to be represented as a row vector while the other a column vector:

```python
vector_one = np.asarray([
    [1],
    [2],
    [1]
], dtype=np.float32)

vector_two = np.asarray([
    [3],
    [0],
    [1]
], dtype=np.float32)

dot_product = np.dot(vector_two[:,0],vector_one) # or np.dot(vector_one[:,0],vector_two)
dot_product
```
